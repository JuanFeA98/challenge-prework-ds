## CHALLENGE-PREWORK-DS

### 1. Importamos las librerias que necesitaremos

```python
import pandas as pd
import numpy as np
```

### 2. Cargamos nuestros archivo csv y lo convertimos en un dataframe

```python
df = pd.read_csv('INAH_detallado_2019.csv', encoding='utf8')
df.head(2)
```

#### Renombramos algunas columnas para que el trabajo sea más fácil

```python
df.rename(columns={
    'Clave SIINAH' : 'clave',
    'Centro de trabajo': 'centro_trabajo',
    'A�o':'year', 
    'Tipo de visitantes':'tipo_visitante',
    'N�mero de visitas':'numero_visitas'}, inplace=True)
```

```python
df.head(2)
```

### 3. Creamos un nuevo dataframe con los datos que vamos a utilizar a lo largo del proyecto

```python
df_work = df.copy(deep=True)
```

```python
df_work.drop(['Siglas'], axis=1, inplace=True)
```

```python
df_work.head(3)
```

#### Observamos el tipo de dato de cada una de las columnas y los ajustamos para que sean el tipo de dato que necesitamos

```python
df_work.dtypes
```

```python
df_work['numero_visitas'] = df_work['numero_visitas'].str.replace(',', '')
```

```python
df_work['numero_visitas']=df_work['numero_visitas'].astype(int)
df_work.dtypes
```

### 4. Empezamos a realizar el análisis de nuestros datos

```python
# Totales por Estado por temporalidad (Mes y Año)

table = pd.pivot_table(df_work, index=['Estado'], values=['numero_visitas'], columns=['year', 'Mes'], aggfunc='sum')

table['totales']=table.sum(axis=1)
table.sample(5)
```

```python
#Totales por Estado por tipo de visitante

table = pd.pivot_table(df_work, index=['Estado'], values=['numero_visitas'], columns=['tipo_visitante'], aggfunc='sum')

table['totales']=table.sum(axis=1)
table.sample(5)
```

```python
#Totales por Estado por tipo de visitante y temporalidad (Mes y Año)

table = pd.pivot_table(df_work, index=['Estado'], values=['numero_visitas'], columns=['year', 'Mes', 'tipo_visitante'], aggfunc='sum')

table['totales']=table.sum(axis=1)
table.sample(5)
```

```python
#Totales por Centro de trabajo y temporalidad (Mes y año)

table = pd.pivot_table(df_work, index=['centro_trabajo'], values=['numero_visitas'], columns=['year', 'Mes'], aggfunc='sum')

table['totales']=table.sum(axis=1)
table
```

```python
#Totales por Centro de trabajo por tipo de visitante (Año)

table = pd.pivot_table(df_work, index=['centro_trabajo'], values=['numero_visitas'], columns=['year', 'tipo_visitante'], aggfunc='sum')
table

table['totales']=table.sum(axis=1)
table
```

```python
#Promedio de visitantes totales por estado (Mes y Año)

table = pd.pivot_table(df_work, index=['Estado'], values=['numero_visitas'], columns=['year', 'Mes'], aggfunc='mean')

pd.options.display.float_format = '{:,.2f}'.format
table
```

```python
#Promedio de visitantes por tipo de visitante, por estado (Mes y Año)

table = pd.pivot_table(df_work, index=['Estado'], values=['numero_visitas'], columns=['year', 'Mes', 'tipo_visitante'], aggfunc='mean')

pd.options.display.float_format = '{:,.2f}'.format

table
```

```python
#Porcentaje de visitantes por Recinto en visitantes por estado (total por mes).

table = pd.pivot_table(df_work, index=['Estado'], values=['numero_visitas'], columns=['year', 'Mes', 'tipo_visitante'], aggfunc='count')

pd.options.display.float_format = '{:,.2f}'.format
table['totales']=table.sum(axis=1)
table
```

```python
#Porcentaje de tipo de visitantes por mes y año
table = pd.pivot_table(df_work, index=['tipo_visitante'], values=['numero_visitas'], columns=['year', 'Mes'], aggfunc='sum')

pd.options.display.float_format = '{:,.2f}'.format
table['totales']=table.sum(axis=1)
table
```

```python
#Porcentaje de tipo de visitantes por recinto (Mes y Año)
table = pd.pivot_table(df_work, index=['Nacionalidad'], values=['numero_visitas'], columns=['year', 'Mes'], aggfunc='sum')

pd.options.display.float_format = '{:,.2f}'.format
table['totales']=table.sum(axis=1)
table
```
